# Introduction

Since Jan. 1, 2015, [The Washington Post](https://www.washingtonpost.com/) has been compiling a database of every fatal shooting in the US by a police officer in the line of duty. 

<center><img src=https://i.imgur.com/sX3K62b.png></center>

While there are many challenges regarding data collection and reporting, The Washington Post has been tracking more than a dozen details about each killing. This includes the race, age and gender of the deceased, whether the person was armed, and whether the victim was experiencing a mental-health crisis. The Washington Post has gathered this supplemental information from law enforcement websites, local new reports, social media, and by monitoring independent databases such as "Killed by police" and "Fatal Encounters". The Post has also conducted additional reporting in many cases.

There are 4 additional datasets: US census data on poverty rate, high school graduation rate, median household income, and racial demographics. [Source of census data](https://factfinder.census.gov/faces/nav/jsf/pages/community_facts.xhtml).

# Chart the Poverty Rate in each US State

Create a bar chart that ranks the poverty rate from highest to lowest by US state. Which state has the highest poverty rate? Which state has the lowest poverty rate?  Bar Plot

```python
df_pct_poverty.poverty_rate = pd.to_numeric(df_pct_poverty.poverty_rate,errors='coerce')

df_pct_poverty = df_pct_poverty.replace(np.nan, 0)
df_pct_poverty.head(575)
poverty_by_state = df_pct_poverty.groupby(['Geographic Area'],dropna=True).mean()
poverty_by_state 
```

```python
# Chart the Poverty Rate in each US State
fig = px.bar(poverty_by_state,
            x=poverty_by_state.index,
            y=poverty_by_state.poverty_rate,
            color = poverty_by_state.index,
            color_continuous_scale=px.colors.sequential.Viridis
            )

fig.update_layout(title="The poverty Rate By US State",
                 xaxis_title= "US State",
                 yaxis_title="% of poverty rate",
                 hovermode="x"
                 )
fig.show()
```

# Chart the High School Graduation Rate by US State

# Visualise the Relationship between Poverty Rates and High School Graduation Rates

#### Create a line chart with two y-axes to show if the rations of poverty and high school graduation move together.  

```python
high_school_completed_rate.values #percent_completed_hs
# col_names =  ['Geographic_Area','percent_completed_hs','poverty_rate']
# relation_poverty_hs = pd.DataFrame(columns = col_names)#(columns=)
# relation_poverty_hs['Geographic_Area'] = poverty_by_state.index#['Geographic Area'] 
# relation_poverty_hs['percent_completed_hs'] = high_school_completed_rate.values#percent_completed_hs#['percent_completed_hs']
# relation_poverty_hs['poverty_rate'] = poverty_by_state.values#['poverty_rate']
# relation_poverty_hs
relation_poverty_hs = poverty_by_state.merge(high_school_completed_rate, how='left', on='Geographic Area')

#relation_poverty_hs.percent_completed_hs = pd.to_numeric(relation_poverty_hs.percent_completed_hs, errors ='coerce')
relation_poverty_hs
```

# Were People Armed? 

In what percentage of police killings were people armed? Create chart that show what kind of weapon (if any) the deceased was carrying. How many of the people killed by police were armed with guns versus unarmed? 

```python
df_killed=[]
df_killed= pd.DataFrame(df_killed,columns=['Number_of_death'])
df_killed['Number_of_death'] = df_fatalities['armed'].value_counts()
df_killed
total_killed = df_killed['Number_of_death'].sum()
total_killed

#the number of ppl killed by the police  who were unarmed
nb_unarmed = df_killed[df_killed.index=='unarmed'].Number_of_death
nb_unarmed
pct_unarmed = (nb_unarmed/ total_killed) *100
pct_unarmed

#the number of ppl killed by the police  who were armed with guns
nb_gun = df_killed[df_killed.index=='gun'].Number_of_death
nb_gun = pd.to_numeric(nb_gun)
pct_gun = (nb_gun/total_killed)*100
print(f'The pourcentage of people killed unarmed is : {pct_unarmed.values}%')
print(f'The pourcentage of armed people with gun is: {pct_gun.values}%')
```

```python
df_fatalities.date = pd.to_datetime(df_fatalities.date,utc=True)
#df_fatalities['year'] =df_fatalities.date(%Y)
dt_index = pd.DatetimeIndex(df_fatalities.date)
year= dt_index.year
df_fatalities['year']=year
df_fatalities
```

```python
killed_over_years = df_fatalities.groupby('year', as_index=False).agg({'id':pd.Series.count})
killed_over_years
```

```python

fig = px.bar(killed_over_years,
            x=killed_over_years.year,
            y=killed_over_years.id,
            color = killed_over_years.year
            )
fig.update_layout(title="Number of Police Killings Over Time",
                  xaxis_title ="Years",
                  yaxis_title = "Number of deaths"
                        )
fig.show()
```
